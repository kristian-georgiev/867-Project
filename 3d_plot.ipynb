```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
grid_x = np.linspace(-1, 1, 100)
grid_y = np.linspace(-1, 1, 100)
loss_grid = np.absolute(np.dot(grid_x.reshape((100,1)), grid_y.reshape((1,100))))

x_traj = [-0.9, 0.5, -0.2, 0.1, 0]
y_traj = [0.95, 0.6, 0.3, 0.2, 0]
```

```python
fig, ax = plt.subplots()

C = ax.contourf(grid_x, grid_y, loss_grid)#, levels=20, cmap=plt.cm.coolwarm)
cbar = fig.colorbar(C)
plt.plot(x_traj, y_traj, c='black', lw='3')
plt.scatter(x_traj[-1], y_traj[-1], marker='X', c='white', s=160)
```

```python
X, Y = np.meshgrid(grid_x, grid_y)
Z = loss_grid

xs = np.array(x_traj)
xs_idx = np.array([np.argmin(np.abs(x-grid_x)) for x in x_traj])

ys = np.array(y_traj)
ys_idx = np.array([np.argmin(np.abs(y-grid_y)) for y in y_traj])

zs = np.array([loss_grid[x_idx][y_idx] for x_idx,y_idx in zip(xs_idx, ys_idx)])

fig = plt.figure()
ax = fig.gca(projection='3d', xlim=(-1,1), ylim=(-1,1), zlim=(-1.5,1.5))

#ax.scatter(xs[-1], ys[-1], zs[-1], marker='X', c='red', s=300, zorder=3)
C = ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False, zorder=4)
cbar=fig.colorbar(C)
ax.plot(xs, ys, zs, c='black', lw='3', zorder=1)
fig.show()
```
